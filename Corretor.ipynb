```python
def separa_palavras(lista_tokens):
  lista_palavras = []
  for token in lista_tokens:
    if token.isalpha():
      lista_palavras.append(token)
  return lista_palavras

```

```python
lista_tokens = nltk.tokenize.word_tokenize(artigos)
lista_palavras = separa_palavras(lista_tokens)
print("O número de palavras é:",len(lista_palavras))
```

esse é o número total de palavras que aparecem no texto porém elas estão repetidas, dessa maneira deve-se normaliza-las

```python
def normalizacao(lista_palavras):
  lista_normalizada = []
  for palavra in lista_palavras:
    lista_normalizada.append(palavra.lower())
  return lista_normalizada

lista_normalizada = normalizacao(lista_palavras)
lista_normalizada
```

```python
import string
pontuacao = string.punctuation

def retira_pontuacao(lista_palavras):
  lista_sem_pontuacao = []
  for palavra in lista_palavras:
    if palavra not in pontuacao:
      lista_sem_pontuacao.append(palavra)

  return lista_sem_pontuacao

lista_sem_pontuacao = retira_pontuacao(lista_palavras)
lista_sem_pontuacao
```

### Construção do modelo

O modelo irá funcionar recebendo uma palavra escrita de forma errada e irá retornar a palavra escrita de maneira certa

Sendo assim iremos inserir letas na palavra

Para que essa operação seja generalizada, iremos pegar cada posição da palavra em questão e inserir caracteres com o resto da palavra ao seu redor, até uma palavra existente ser formada. Isso pode cobrir um erro cometido em qualquer posição da palavra

```python
def insere_letras(fatias):
  novas_palavras = []
  letras = 'abcdefghijklmnopqrstuvwxyzàáâãèéêìíîòóôõùúûç'
  for E,D in fatias:
    for letra in letras:
      novas_palavras.append(E + letra + D)
  return novas_palavras
```

```python
palavra_exemplo = 'proocuar'
def gerador_palavras(palavra):
  fatias = []
  for i in range(len(palavra)+1):
    fatias.append((palavra[:i], palavra[i:]))
  palavras_geradas = insere_letras(fatias)
  return palavras_geradas
```

Para criar o gerador vamos ver qual é combinação de maior probabilidade de ser a palavra certa

Na função max podemos chamar uma outra função (vamos criar uma de probabilidade que irá retornar a probabilidade de cada variação de palavra, a max irá apenas tirar a maior delas)

```python
total_palavras = len(lista_sem_pontuacao)
frequencia = nltk.FreqDist(lista_sem_pontuacao) #irá registrar a frequência de cada palavra do artigo
```

```python
def probabilidade(palavra):
  return frequencia[palavra]/total_palavras
```

```python
def corretor(palavra):
  palavras_geradas = gerador_palavras(palavra)
  palavra_correta = max(palavras_geradas, key = probabilidade)
  return palavra_correta

```

O gerador turbinado irá receber palavras já geradas nas funções anteriores

```python
def gerador_turbinado(palavras_geradas):
  novas_palavras = []
  for palavra in palavras_geradas:
    novas_palavras += gerador_palavras(palavra)
  return novas_palavras

palavras_g = gerador_turbinado(gerador_palavras("lóiigica"))
"lógica" in palavras_g
```

Para resolver o problema, vamos modificar a função corretor, ela irá retornar uma nova lista apenas com a palavra a ser corrigida e as palavras existentes no vocabulário

```python
def novo_corretor(palavra):
  palavras_geradas = gerador_palavras(palavra)
  palavras_geradas_turbinado = gerador_turbinado(palavras_geradas)
  todas_palavras = set(palavras_geradas + palavras_geradas_turbinado)
  candidatos = [palavra]
  for palavra in todas_palavras:
    if palavra in vocabulario:
      candidatos.append(palavra)
  palavra_correta = max(candidatos, key = probabilidade)
  return palavra_correta

novo_corretor("lóiigica")
```
